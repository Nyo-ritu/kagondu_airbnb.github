```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

print(air_bnb.nunique())
#There are 5 neighborhood groups.
print('\n')
print(air_bnb.neighbourhood_group.mode())
#Manhatten shows up the most.
```

```python
# Are private rooms the most popular in manhattan?
pop_rooms = air_bnb['room_type'].mode()
print(pop_rooms)

#NO, the most popular roomtype is the entire home/apt
```

```python
# Which hosts are the busiest and based on their reviews?

busy_host = air_bnb.sort_values('number_of_reviews', kind = 'mergesort', ascending = False).head()
busy_host

#top 5 hosts with the most reviews listed below. no.1 is Dona, with 629 reviews.
```

```python
#Which neighorhood group has the highest average price? Without checking anything, the answer is obviously Manhattan :)
high_av_price = air_bnb.groupby('neighbourhood_group').mean()['price']
high_av_price

#Manhattan has the highest average price. $196.87 (per night I assume)
```

```python
# Which neighborhood group has the highest total price?
high_total_price = air_bnb.groupby('neighbourhood_group').sum()['price']
high_total_price

#Manhattan has the highest total price, with a combined sum of $4,264,527.
```

```python
#Which top 5 hosts have the highest total price?
high_total_price = air_bnb.sort_values('price', kind = 'mergesort', ascending = False).head()
high_total_price

#Top 5 most expensive hosts are Kathrine in Queens, Erin in Brooklyn, Jelena in Manhatan,
#Olson in Manhattan and Amy in Manhattan.
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

no_avail = air_bnb[air_bnb['availability_365'] == 0]

no_availl = no_avail[no_avail['number_of_reviews'] >= 100]
no_availl

#There are 162 hosts with zero availability and a review count of 100 or more
```

```python
# What host has the highest total of prices and where are they located?

highest_prices = air_bnb.sort_values('price', kind = 'mergesort', ascending = False).head().max()


highest_prices

#Olsen in Queens
```

```python
# When did Danielle from Queens last receive a review?

Dan = air_bnb[air_bnb.host_name == 'Danielle']  #count()
Dann = Dan[Dan.neighbourhood_group == 'Queens'].last_review #count()
Dann
# There are 76 hosts called Danielle, 8 are in Queens, 7 have a last review, listed below. 
```
